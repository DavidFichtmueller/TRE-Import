```python
#define the import function

from wikibaseintegrator.datatypes import ExternalID, Item, String, Time
from wikibaseintegrator.wbi_exceptions import ModificationFailed

def importTaxonName(taxonNames,wikibase_login):
    

    wbi = WikibaseIntegrator(login=wikibase_login)

    count = 0
    start=time.time()


    for index, taxonName in taxonNames.iterrows():

        # Create a new item
        item = wbi.item.new()

        # Set an English label
        label = taxonName.canonicalName
        if taxonName.canonicalName == "":
            label = taxonName.scientificName

        item.labels.set(language='en', value=label)

        # Set a English description
        addendum = ""
        if(taxonName.taxonomicStatus != "accepted"):
            addendum = "("+taxonName.taxonomicStatus+")"
        
        #Special rule for some homonyms that cause issues when imported
        #TODO rewrite so it works for all of the taxon ranks by always selecting the next higher one 
        #addendum = addendum + " from the order "+taxonName.order
        addendum = addendum.strip()
        item.descriptions.set(language='en', value="a taxonomic name of rank "+taxonName.taxonRank+" "+addendum)

        statements = []
        # data type object
        statements.append(Item(value='Q1', prop_nr='P1')) #instance of: Taxon Name
        statements.append(Item(value='Q9', prop_nr='P8')) #source: GBIF Taxonomic Backbone
        statements.append(String(value=taxonName.taxonID, prop_nr='P7'))
        #map the manually created taxon ranks
        taxonRanks = {"kingdom":"Q10", "phylum":"Q11", "class":"Q12", "order":"Q13", "family":"Q14", "genus":"Q15", "species":"Q16", "subspecies":"Q17", "variety":"Q18", "form":"Q19", "unranked":"Q20" }
        statements.append(Item(value=taxonRanks[taxonName.taxonRank], prop_nr='P5'))
        #mal the manually created taxon statuses
        taxonomicStatus = {"accepted":"Q22", "doubtful":"Q23", "synonym":"Q24", "homotypic synonym":"Q25", "heterotypic synonym":"Q26", "proparte synonym":"Q27" }
        statements.append(Item(value=taxonomicStatus[taxonName.taxonomicStatus], prop_nr='P13'))
        if taxonName.scientificName != "":
            statements.append(String(value=taxonName.scientificName, prop_nr='P14'))
        if taxonName.scientificNameAuthorship != "":
            statements.append(String(value=taxonName.scientificNameAuthorship, prop_nr='P15'))
        if taxonName.kingdom != "":
            statements.append(String(value=taxonName.kingdom, prop_nr='P16'))
        if taxonName.phylum != "":
            statements.append(String(value=taxonName.phylum, prop_nr='P17'))
        if taxonName["class"] != "":
            statements.append(String(value=taxonName["class"], prop_nr='P18'))
        if taxonName.order != "":
            statements.append(String(value=taxonName.order, prop_nr='P19'))
        if taxonName.family != "":
            statements.append(String(value=taxonName.family, prop_nr='P20'))
        if taxonName.genus != "":
            statements.append(String(value=taxonName.genus, prop_nr='P21'))
        if taxonName.specificEpithet != "":
            statements.append(String(value=taxonName.specificEpithet, prop_nr='P22'))
        if taxonName.infraspecificEpithet != "":
            statements.append(String(value=taxonName.infraspecificEpithet, prop_nr='P23'))

        item.claims.add(statements)
        try:
            #DISABLED FOR SAFETY REASONS
            item.write(tags = ['wbi','import-names'])
        except ModificationFailed as error:
            print(error.info)
            #print("error")
            e = error
            
        
        #throttle editing rate, to not overwhelm the Glazegraph injestion
        time.sleep(0.1)
        
        count = count +1
        if count%max(round(len(taxonNames)/100),1) == 0 :
            end=time.time()
            print(str(round(count*100/len(taxonNames))) + "%: "+str(round((end-start)*100)/100)+" seconds")
```

```python
#query the mappings from the Q-IDs to the GBIF-IDs
from wikibaseintegrator import wbi_helpers

query = """
SELECT ?taxonNameItem ?GBIF_ID WHERE {
?taxonNameItem <https://tre-test.wikibase.cloud/prop/direct/P7> ?GBIF_ID. 
}
"""

qid_gbifid_map = wbi_helpers.execute_sparql_query(query)
qid_gbifid_map = pd.json_normalize(qid_gbifid_map['results']['bindings'])
qid_gbifid_map = qid_gbifid_map[['taxonNameItem.value', 'GBIF_ID.value']]
qid_gbifid_map.columns = ['taxonNameItem', 'GBIF_ID']
qid_gbifid_map['taxonNameItem'] = qid_gbifid_map['taxonNameItem'].str.replace('https://tre-test.wikibase.cloud/entity/','')
qid_gbifid_map
```

```python
#merge the two tables as a left join
accepted_bees_with_qids = pd.merge(accepted_bees, qid_gbifid_map, left_on="taxonID", right_on="GBIF_ID", how="left", sort=False)
#drop column "GBIF_ID" as it is the same as the "taxonID" 
accepted_bees_with_qids = accepted_bees_with_qids.drop(columns = ['GBIF_ID'])
#rename the column "taxonNameItem" to qid
accepted_bees_with_qids = accepted_bees_with_qids.rename(columns={'taxonNameItem': 'qid'})
#move column "qid" to the second postion
col = accepted_bees_with_qids.pop("qid")
accepted_bees_with_qids.insert(1, col.name, col)


#merge the two tables again as a left join, thistime for the "parentNameUsageID"
accepted_bees_with_qids = pd.merge(accepted_bees_with_qids, qid_gbifid_map, left_on="parentNameUsageID", right_on="GBIF_ID", how="left", sort=False)
#drop column "GBIF_ID" as it is the same as the "parentNameUsageID" 
accepted_bees_with_qids = accepted_bees_with_qids.drop(columns = ['GBIF_ID'])
#rename the column "taxonNameItem" to qid
accepted_bees_with_qids = accepted_bees_with_qids.rename(columns={'taxonNameItem': 'parentQid'})
#move column "qid" to the fifths postion
col = accepted_bees_with_qids.pop("parentQid")
accepted_bees_with_qids.insert(4, col.name, col)

accepted_bees_with_qids.qid = accepted_bees_with_qids.qid.fillna('')
accepted_bees_with_qids.parentQid = accepted_bees_with_qids.parentQid.fillna('')
accepted_bees_with_qids
```

```python
def importTaxonNameParentLink(taxonNames,wikibase_login):
    

    wbi = WikibaseIntegrator(login=wikibase_login)

    count = 0
    start=time.time()


    for index, taxonName in taxonNames.iterrows():
        if taxonName.parentQid == "":
            continue
        item = wbi.item.get(entity_id=taxonName.qid)
        count = count+1
        if("P6" in item.get_json()["claims"]):
            continue
        statements = []
        statements.append(Item(value=taxonName.parentQid, prop_nr='P6')) #parentTaxonName
        item.claims.add(statements)
        
        #DISABLED FOR SAFETY REASONS
        #item.write(tags = ['wbi','import-names'])
        time.sleep(0.2)

        if count%round(len(taxonNames)/100) == 0 :
            end=time.time()
            print(str(round(count*100/len(taxonNames))) + "%: "+str(round((end-start)*100)/100)+" seconds")
```

```python
#query the mappings from the Q-IDs to the GBIF-IDs
from wikibaseintegrator import wbi_helpers

query = """
PREFIX tre: <https://tre-test.wikibase.cloud/entity/>
PREFIX tred: <https://tre-test.wikibase.cloud/prop/direct/>

SELECT ?taxonName ?id WHERE {
  ?taxonName tred:P7 ?id.
  ?taxonName tred:P5 tre:Q14.
}ORDER BY ?taxonName
"""

qid_gbifid_map_family = wbi_helpers.execute_sparql_query(query)
qid_gbifid_map_family = pd.json_normalize(qid_gbifid_map_family['results']['bindings'])
qid_gbifid_map_family = qid_gbifid_map_family[['taxonName.value', 'id.value']]
qid_gbifid_map_family.columns = ['taxonNameItem', 'GBIF_ID']
qid_gbifid_map_family['taxonNameItem'] = qid_gbifid_map_family['taxonNameItem'].str.replace('https://tre-test.wikibase.cloud/entity/','')
qid_gbifid_map_family
```

```python
#check for missing items by iterating over the range of ids that were generated during the import
#this assumes that no other imports or item creations were run during the run of the import
#if an id does not appear in the results of the previous query, there are two possibilites: 
# 1. there is no item at that ID, which is the case, when an error was thrown during the item creation, most commonly
#    because an already existing item already has the same label and description in one langauge
# 2. because there was an issue with updating the query service. The item was created correctly but it doesn't show 
#    up in the results. In this case, we get the item information via the API, query the GBIF Taxon ID that belongs 
#    to the item and add it to the list

from wikibaseintegrator.wbi_exceptions import MissingEntityException

for i in range(13304,47295):
    qid = "Q"+str(i)
    if(len(qid_gbifid_map_family[qid_gbifid_map_family["taxonNameItem"] == qid])==0):
        try:
            missing_item = wbi.item.get(entity_id=qid)
            missing_id = missing_item.get_json()["claims"]["P7"][0]["mainsnak"]["datavalue"]["value"]
            df_tmp = pd.DataFrame([[qid,missing_id]], columns=['taxonNameItem','GBIF_ID'])
            qid_gbifid_map_family = pd.concat([qid_gbifid_map_family, df_tmp]).reset_index(drop=True)
        except MissingEntityException as error:
            print("Missing: "+qid)
        
        
```
